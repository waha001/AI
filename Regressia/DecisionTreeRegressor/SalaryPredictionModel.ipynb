```python
#Kerakli bolgan barcha kutubxonalarni yuklab olamiz
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
# Bu yerda odiy lug'at yaratganmz
data = {
    'Yosh': [25, 30, 35, 40, 45, 50, 55, 60],
    'Tajriba': [1, 3, 5, 7, 9, 11, 13, 15],
    'Maosh': [300, 450, 600, 750, 900, 1050, 1200, 1350]
}
```

```python
# bu qadamda tepadagi lugatimni pandasda dataframega o'zgartiramiz
df = pd.DataFrame(data)
```

```python
# Bu qadamda datamizni x va y ga ajratamiz
X = df[['Yosh', 'Tajriba']]
y = df['Maosh']
```

```python
# Modellangan daraxtni vizual koâ€˜rsat va Qaror daraxti regressiya modelini yaratish uchun kutubxona.
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
# Madel yaratdik, max_depth=1 madelni qanchalik chuqur o'rganishini belgilaymiz
tree = DecisionTreeRegressor(max_depth=1)
```

```python
# Madelga o'rgatdik
tree.fit(X, y)
```

```python
# bashorat qilish uchun yangi data tuzdik
new= pd.DataFrame({'Yosh': [20],'Tajriba': [1] })
```

```python
# Datamizni bashorat qilish uchun berdik
predicted_sales = tree.predict(new)
```

```python
# Bashorat natijasini oynaga chiqardik
print("Natija:", predicted_sales[0])
```

```python
# Grafik o'lchamini belgilaymiz
plt.figure(figsize=(8, 4))
# Qaror daraxtini vizual ko'rinishda chiqaradi
plot_tree(tree, feature_names=X.columns, filled=True, rounded=True)
plt.title("Decision Tree ")
plt.show()
```
